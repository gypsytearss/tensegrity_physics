```python
xs = np.linspace(0,4*np.pi,400)
y_sin = map(lambda x: np.sin(x), xs)
y_cos = map(lambda x: np.cos(x), xs)
y_sums = map(lambda x: np.sin(x) - np.cos(x), xs)

fig = plt.figure()
plt.plot(xs, y_sin)
plt.plot(xs, y_cos)
plt.plot(xs, y_sums)
plt.show()
```
